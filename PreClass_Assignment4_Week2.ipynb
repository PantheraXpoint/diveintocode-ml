```python
import numpy as np
inp = input("Enter the times of throwing darts:")
dice = np.random.choice(5,1)
dice += 1
coordinate = []
for i in range(int(inp)):
    x = np.random.normal(dice,0.3,(1,1))
    y = np.random.normal(0,0.3,(1,1))
    coordinate.append((x,y))
ret = dict.fromkeys(range(int(inp)),coordinate)  
for i,j in ret.items():
    print("The coordinate of the dart after " + str(i) + " throw(s):")
    print(j[i])
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 1000)

r1 = np.sqrt(1.0)
r2 = np.sqrt(2/3)
r3 = np.sqrt(1/3)
x1 = r1*np.cos(theta) + dice
x2 = r1*np.sin(theta)
x3 = r2*np.cos(theta) + dice
x4 = r2*np.sin(theta)
x5 = r3*np.cos(theta) + dice
x6 = r3*np.sin(theta)

fig, ax = plt.subplots(1)

ax.plot(x1, x2,x3,x4,x5,x6)
ax.set_aspect(1)
for i,j in ret.items():
    plt.plot(j[i][0],j[i][1],'bo',markersize=1) 





plt.title('How to plot a circle with matplotlib ?', fontsize=8)

plt.savefig("plot_circle_matplotlib_01.png", bbox_inches='tight')

plt.show()
ax.set_xlim(0, 6.5)
ax.set_ylim(-0.5, 0.5)
ax.set_title('DART THROWING SIMULATION')
ax.set_aspect(1.0)
plt.show()
```
