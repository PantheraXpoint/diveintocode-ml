```python
import numpy as np
import matplotlib.pyplot as plt
mean_p4 = (0,-3)
cov = [[1, 0.8], [0.8, 1]]
np.random.seed(0)
B = np.random.multivariate_normal(mean_p4,cov,(500,2))
X_B = B[:,0]
mean_p1 = (-3,0)
np.random.seed(0)
A = np.random.multivariate_normal(mean_p1,cov,(500,2))
X_A = A[:,0]
#plt.plot(A, label = "0")
#plt.plot(B, label = "1")
#plt.legend(loc = 'best')
Z = np.concatenate((X_A,X_B), axis = 0)
N = np.array([0])
for i in range(1,len(X_A)):
    N = np.append(N,0)
for i in range(len(X_A), len(Z)):
    N = np.append(N,1)
ret = np.concatenate((Z,N[:,None]), axis = 1)
print(ret)
```
