```python
import matplotlib.pyplot as plt
import numpy as np

def func_1(x = np.arange(-50,50.1,0.1)):
    return x,len(x),x * x

def func_2(x = np.arange(-50,50.1,0.1)):
    return x,len(x),2 * x * x + 2**x
    
def func_3(x = np.arange(0,50.1,0.1)):
    return x,len(x), np.sin(x**(1/2))

def gradient(func):
    x,length,y = func 
    XY = np.stack((x,y),axis = 1)
    d_x = np.array(XY[1:,:1] - XY[:-1,:1])
    d_y = np.array(XY[1:,1:] - XY[:-1,1:])
    grad = np.array(d_y / d_x)
    return x,XY,grad

x,arr_xy1, grad1 = gradient(func_1())
x,arr_xy2, grad2 = gradient(func_2())
x_,arr_xy3, grad3 = gradient(func_3())
x = np.arange(-50,50,0.1)
x_ = np.arange(0,50,0.1)
plt.plot(x,grad1)
plt.xlabel('x')
plt.ylabel('gradient')
plt.show()

plt.plot(x,grad2)
plt.xlabel('x')
plt.ylabel('gradient')
plt.show()

plt.plot(x_,grad3)
plt.xlabel('x')
plt.ylabel('gradient')
plt.show()
    
    
    
```
