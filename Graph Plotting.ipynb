```python
X = np.linspace(0,2,10)
y = X**2

plt.plot(X,y)
plt.title("figure 1")
plt.xlabel("axis x")
plt.ylabel("axis y")
plt.grid(axis = 'y')
plt.show()
```

```python
X = np.linspace(0,2,10)

plt.figure(figsize=(12,6)) #Size of the figure

plt.plot(X,X**2,label = 'First Graph')
plt.plot(X,X**3, label = 'Second Graph')

#Extra

plt.title("figure 1", loc = 'left') #title on the left
plt.xlabel("axis x")
plt.ylabel("axis y")
plt.legend()

plt.savefig("test_figure.png")
plt.show()
```

```python
x = np.linspace(0,2,10)
y = x**2
plt.subplot(2,2,1)
plt.plot(x,y,lw = 4,c='red')
plt.subplot(2,2,2)
plt.plot(x,y,lw=4,c='blue')
plt.subplot(2,2,3)
plt.plot(x,y,lw = 4,c='red')
plt.subplot(2,2,4)
plt.plot(x,y,lw=4,c='blue')
plt.suptitle("Title")
```

```python
fig,ax = plt.subplots(1,2,sharex = True) #share the same axis for subplots
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.show()
```

```python
x = np.arange(-50,50.1,0.1)
x2 = np.arange(0,50.1,0.1)

y1 = np.array([(1/2) * i + 1 for i in x])
y2 = x**2
y3 = (2 * (x**2) + (2**x))
y4 = np.sin(x2**(1/2))

my_figure = plt.subplots(nrows = 2, ncols =2, figsize=(14,6))
fig,((ax1,ax2),(ax3,ax4)) = my_figure

ax1.plot(y1,c='crimson')
ax2.plot(y2,c='black')
ax3.plot(y3,c='blue')
ax4.plot(y4,c='purple')

plt.suptitle("Maths functions")
plt.show()
```
