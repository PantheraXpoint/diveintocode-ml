```python
import numpy as np
bun_radius = 0.02     # the radius of the chestnut bun
solar_radius = 287460000000000 # As I surf on the internet, the solar system size has some invisible part so my figure might be different to others.
bun_area = bun_radius**2 * np.pi          # As I cannot estimate the volume of the solar system, so I measure the area instead
solar_area = solar_radius**2 * np.pi      # Imagine the solar system as a plat disk
doubling_times = 0   # number of times the buns double themselves
while (bun_area < solar_area):
    bun_area *= 2
    doubling_times += 1
how_long = (doubling_times * 5)
print("It takes {} minutes to cover the solar system\n".format(how_long))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
def calculation(sprinkle_obj_volume,fillup_obj_volume):
    sprinkle_obj = []
    fillup_obj = []
    doubling_times = 0      # number of times the object double its size
    true_time = []          # timeline 0,5,10,15,..... as 5 minutes the object double its size
    true_time.append(doubling_times)
    sprinkle_obj.append(sprinkle_obj_volume) # update the total sprinkled objects' volume after x times
    fillup_obj.append(fillup_obj_volume)     #the constant value, however needed for ploting later on
    sprinkle = sprinkle_obj_volume           #variable for storing and calculating
    while (sprinkle < fillup_obj_volume):
        sprinkle *= 2
        doubling_times += 1
        true_time.append(doubling_times * 5)
        fillup_obj.append(fillup_obj_volume)
        sprinkle_obj.append(sprinkle)
    return fillup_obj,sprinkle_obj,doubling_times,true_time

#store the increasing volume of the sprinkled objects and the volume of the constant object

sprinkle_obj_volume = float(input("Enter the volume of the sprinkled objects: (in cubic meters): "))
fillup_obj_volume = float(input("Enter the radius of the object you want to fill up: (in cubic meters): "))
fillup_obj,sprinkle_obj,doubling_times,true_time = calculation(sprinkle_obj_volume,fillup_obj_volume)
how_long = (doubling_times * 5)

# create a line graph to see how long the sprinkled objects surpass the object that needs filling up according to time(minutes)

plt.xticks(sprinkle_obj,true_time)
plt.plot(sprinkle_obj, label = "Sprinkled Object's Volume")
plt.plot(fillup_obj, label = "Given Object's Volume")
plt.legend(loc = 'best')
plt.title("How Long Will It Take For A Sprinkled Object's Volume To Reach The Given Object's Volume\n")
plt.ylabel("Volume(cubic meters)")
plt.xlabel("time(minutes)")

plt.show()
print("It takes {} minutes to cover the given object\n".format(how_long))
```
